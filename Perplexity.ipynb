## **Perplexity 20N**

```python
import math

#Calculate perplexity of the unigram model

def unigram_perplexity():

    unigrams_probabilities = {}

    with open('20N_unigrams.txt','r') as file:
        for line in file:
            line = line.strip()
            modified_string = line.replace(":", "").replace("'","")
            list = modified_string.split()
            unigrams_probabilities[list[0]] = list[1]
            
    corpus = []
    with open('20N_testing.txt', 'r') as file:
        for line in file:
            line = line.strip()
            list = line.replace("'","").split()
            corpus.append(list)
    test_data = [item for sublist in corpus for item in sublist] #List of words separated

    N = len(test_data)
    log_prob_sum = 0
    
    for word in test_data:
        probability = float(unigrams_probabilities.get(word,unigrams_probabilities["<UNK>"]))
        log_prob_sum += math.log((probability))
    
    perplexity = math.exp(-log_prob_sum / N)


    return perplexity

print("The perplexity of the 20N unigrams is: " + str(unigram_perplexity()))
```

```python
#Calculate perplexity of the bigram model
from collections import Counter

def bigram_perplexity():

    bigrams_probabilities = {}

    with open('20N_bigrams.txt','r') as file:
        big = []
        for line in file:
            line = line.strip()
            modified_string = line.replace("'","")
            list = modified_string.split(":")
            big.append(list)
            bigrams_probabilities[list[0]] = list[1]
            
    corpus = []
    with open('20N_testing.txt', 'r') as file:
        for line in file:
            line = line.strip()
            list = line.replace("'","").split()
            corpus.append(list)
    test_data = [item for sublist in corpus for item in sublist] #List of words separated

    bigram_counts = Counter(test_data)
    
    V = len(bigram_counts) # Vocabulary

    N = len(test_data) - 1  # Total number of bigrams
    
    log_prob_sum = 0

    for i in range(1, len(test_data)):
        prev_word = test_data[i - 1]
        current_word = test_data[i]
        probability = float(bigrams_probabilities.get(f'({prev_word}, {current_word})',1/V))

        log_prob_sum += math.log(probability)

    perplexity = math.exp(-log_prob_sum / N)
    
    return perplexity

bigram = bigram_perplexity()
    
print("The perplexity of the 20N bigrams is: " + str(bigram))
```
